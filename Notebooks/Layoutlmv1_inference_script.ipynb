### Feature Extractor

```python
import collections
from transformers import LayoutLMTokenizer

tokenizer = LayoutLMTokenizer.from_pretrained("microsoft/layoutlm-base-uncased")

args = {'local_rank': -1,
        'overwrite_cache': True,
        'data_dir': '/content/drive/MyDrive/data/',
        'model_name_or_path':'microsoft/layoutlm-base-uncased',
        'max_seq_length': 512,
        'model_type': 'layoutlm','doc_stride': 384}

# class to turn the keys of a dict into attributes (thanks Stackoverflow)
class AttrDict(dict):
    def __init__(self, *args, **kwargs):
        super(AttrDict, self).__init__(*args, **kwargs)
        self.__dict__ = self

args = AttrDict(args)

def _check_is_max_context(doc_spans, cur_span_index, position):
    """Check if this is the 'max context' doc span for the token."""

    # Because of the sliding window approach taken to scoring documents, a single
    # token can appear in multiple documents. E.g.
    #  Doc: the man went to the store and bought a gallon of milk
    #  Span A: the man went to the
    #  Span B: to the store and bought
    #  Span C: and bought a gallon of
    #  ...
    #
    # Now the word 'bought' will have two scores from spans B and C. We only
    # want to consider the score with "maximum context", which we define as
    # the *minimum* of its left and right context (the *sum* of left and
    # right context will always be the same, of course).
    #
    # In the example the maximum context for 'bought' would be span C since
    # it has 1 left context and 3 right context, while span B has 4 left context
    # and 0 right context.
    best_score = None
    best_span_index = None
    for (span_index, doc_span) in enumerate(doc_spans):
        end = doc_span.start + doc_span.length - 1
        if position < doc_span.start:
            continue
        if position > end:
            continue
        num_left_context = position - doc_span.start
        num_right_context = end - position
        score = min(num_left_context, num_right_context) + 0.01 * doc_span.length
        if best_score is None or score > best_score:
            best_score = score
            best_span_index = span_index

    return cur_span_index == best_span_index


class InputFeatures(object):
    """A single set of features of data."""

    def __init__(
        self,
        tokens,
        input_ids,
        input_mask,
        segment_ids,
        boxes,
        actual_bboxes,
        token_actual_boxes,
        token_is_max_context,
        doc_span_index,
        example_index,
        text_segment_idx,
        original_token_to_word,
        ocr_confidence
    ):
        assert (
            0 <= all(boxes) <= 1000
        ), "Error with input bbox ({}): the coordinate value is not between 0 and 1000".format(
            boxes
        )
        self.tokens = tokens
        self.input_ids = input_ids
        self.input_mask = input_mask
        self.segment_ids = segment_ids
        self.boxes = boxes
        self.actual_bboxes = actual_bboxes
        self.token_actual_boxes = token_actual_boxes
        self.token_is_max_context = token_is_max_context
        self.doc_span_index = doc_span_index
        self.example_index = example_index
        self.text_segment_idx = text_segment_idx
        self.original_token_to_word= original_token_to_word
        self.ocr_confidence = ocr_confidence


def convert_single_example_to_features(ex_index, image, words, boxes, actual_boxes, ocr_confidence_scores, tokenizer, args, cls_token_box=[0, 0, 0, 0],
                                 sep_token_box=[1000, 1000, 1000, 1000],
                                 pad_token_box=[0, 0, 0, 0]):
      width, height = image.size

      tokens = []
      token_boxes = []
      actual_bboxes = [] # we use an extra b because actual_boxes is already used
      token_actual_boxes = []
      original_token = []
      ocr_confidence_scores_list = []
      for word, box, actual_bbox, ocr_conf in zip(words, boxes, actual_boxes, ocr_confidence_scores):
          word_tokens = tokenizer.tokenize(word)
          original_token.extend([(word, word_token) for word_token in word_tokens])
          tokens.extend(word_tokens)
          token_boxes.extend([box] * len(word_tokens))
          ocr_confidence_scores_list.extend([ocr_conf] * len(word_tokens))
          actual_bboxes.extend([actual_bbox] * len(word_tokens))
          token_actual_boxes.extend([actual_bbox] * len(word_tokens))

      # print(original_token)
      # Truncation: account for [CLS] and [SEP] with "- 2".
      special_tokens_count = 2

      # print(">>>>>>>>>>>>>>>>", len(tokens))
      max_tokens_for_doc = args.max_seq_length - 2
      # max_tokens_for_doc = max_seq_length - len(tokens) - 2
      _DocSpan = collections.namedtuple(  # pylint: disable=invalid-name
          "DocSpan", ["start", "length"])
      doc_spans = []
      start_offset = 0
      while start_offset < len(tokens):
          length = len(tokens) - start_offset
          if length > max_tokens_for_doc:
              length = max_tokens_for_doc
          doc_spans.append(_DocSpan(start=start_offset, length=length))
          if start_offset + length == len(tokens):
              break
          start_offset += min(length, args.doc_stride)

      feature_list, ntokens_list, token_actual_boxes_list, actual_boxes_list, token_boxes_list = [], [], [], [], []
      original_token_to_word_list = []
      segment = 0
      for (doc_span_index, doc_span) in enumerate(doc_spans):

          token_is_max_context = {}
          token = []
          token_box = []
          actual_bbox = []
          token_actual_box = []
          segment_id = []
          original_token_to_word = []
          ocr_conf_score = []

          # print(doc_span.length, doc_spans)
          for i in range(doc_span.length):
              split_token_index = doc_span.start + i
              # token_to_orig_map[len(ntokens)] = tok_to_orig_index[split_token_index]

              is_max_context = _check_is_max_context(doc_spans, doc_span_index,
                                                    split_token_index)
              token_is_max_context[split_token_index] = is_max_context
              token.append(tokens[split_token_index])
              original_token_to_word.append(original_token[split_token_index])
              token_box.append(token_boxes[split_token_index])
              actual_bbox.append(actual_bboxes[split_token_index])
              token_actual_box.append(token_actual_boxes[split_token_index])
              ocr_conf_score.append(ocr_confidence_scores_list[split_token_index])
              # label_id.append(label_map[label_ids[split_token_index]])
              segment_id.append(0)
          # print(len(original_token_to_word), len(token))
          # add [SEP] token, with corresponding token boxes and actual boxes
          token += [tokenizer.sep_token]
          token_box += [sep_token_box]
          actual_bbox += [[0, 0, width, height]]
          token_actual_box += [[0, 0, width, height]]
          original_token_to_word +=[tokenizer.sep_token]
          ocr_conf_score += [0.0]

          segment_id = [0] * len(token)

          # next: [CLS] token
          token = [tokenizer.cls_token] + token
          token_box = [cls_token_box] + token_box
          actual_bbox = [[0, 0, width, height]] + actual_bbox
          token_actual_box = [[0, 0, width, height]] + token_actual_box
          segment_id = [1] + segment_id
          original_token_to_word = [tokenizer.cls_token] + original_token_to_word
          ocr_conf_score = [0.0] + ocr_conf_score

          input_id = tokenizer.convert_tokens_to_ids(token)

          # The mask has 1 for real tokens and 0 for padding tokens. Only real
          # tokens are attended to.
          input_mask = [1] * len(input_id)

          # Zero-pad up to the sequence length.
          padding_length = args.max_seq_length - len(input_id)
          input_id += [tokenizer.pad_token_id] * padding_length
          input_mask += [0] * padding_length
          segment_id += [tokenizer.pad_token_id] * padding_length
          token_box += [pad_token_box] * padding_length
          token_actual_box += [pad_token_box] * padding_length
          text_segment_idx = [segment] * len(input_id)
          ocr_conf_score += [0.0] * padding_length
          # print(text_segment_idx)

          assert len(input_id) == args.max_seq_length
          assert len(input_mask) == args.max_seq_length
          assert len(segment_id) == args.max_seq_length
          assert len(text_segment_idx) == args.max_seq_length
          assert len(token_box) == args.max_seq_length
          assert len(token_actual_box) == args.max_seq_length
          assert len(ocr_conf_score) == args.max_seq_length

          feature = InputFeatures(
                    tokens=token,
                    input_ids=input_id,
                    input_mask=input_mask,
                    segment_ids=segment_id,
                    boxes=token_box,
                    actual_bboxes=actual_bbox,
                    token_actual_boxes=token_actual_box,
                    token_is_max_context=token_is_max_context,
                    doc_span_index=doc_span_index,
                    example_index=ex_index,
                    original_token_to_word=original_token_to_word,
                    text_segment_idx=text_segment_idx,
                    ocr_confidence=ocr_conf_score
                )

          feature_list.append(feature)
          ntokens_list.append(token)
          token_actual_boxes_list.append(token_actual_box)
          actual_boxes_list.append(actual_bbox)
          token_boxes_list.append(token_box)
          segment = segment + 1

      return feature_list, ntokens_list, token_actual_boxes_list, actual_boxes_list, token_boxes_list, original_token
```

```python
def convert_to_features(images_list, words_list, boxes_list, actual_boxes_list, confidence_list, tokenizer, args):
    batch_tokens = []
    batch_token_actual_boxes = []
    batch_tokens_boxes = []
    batch_actual_bboxes = []
    batch_index = []
    feature_list_total = []
    original_token_list = []
    for ex_index, (image, words, boxes, actual_boxes, ocr_score) in enumerate(zip(images_list, words_list, boxes_list, actual_boxes_list, confidence_list)):
        feature_list, ntokens_list, token_actual_boxes_list, actual_boxes_list, token_boxes_list, original_token = convert_single_example_to_features(ex_index, image=image, words=words, boxes=boxes, actual_boxes=actual_boxes, ocr_confidence_scores=ocr_score, tokenizer=tokenizer, args=args)
        # ex_index = 0
        feature_list_total.extend(feature_list)
        original_token_list.append(original_token)
        for feature, ntokens, token_actual_boxes, actual_boxes, token_boxes in zip(feature_list, ntokens_list, token_actual_boxes_list, actual_boxes_list, token_boxes_list):
            batch_tokens.extend(ntokens)
            batch_token_actual_boxes.extend(token_actual_boxes)
            batch_index.extend([ex_index]*len(ntokens))
            batch_tokens_boxes.extend(token_boxes)
            batch_actual_bboxes.extend(actual_boxes)

    return feature_list_total, batch_tokens, batch_token_actual_boxes, batch_index, batch_tokens_boxes, batch_actual_bboxes, original_token_list
```

```python
# pred_tup_list
def remove_rows_after_last_match(df, column_name):
    try:
        # LOGGER.info("remove_rows_after_last_match- ignoring last few rows to avoid noise")
        count = 0
        total_rows = len(df)
        end_index = total_rows
        for index, row in df.iloc[:end_index][::-1].iterrows():
            if row[column_name].startswith('E-') or row[column_name].startswith('S-'):
                # df.loc[index+1:].to_csv(f'./remove_rows_after_last_match_{index}.csv')
                df = df.loc[:index+1].reset_index(inplace = False)
                break
            else:
                count = count + 1
        return df
    except Exception as e:
        # LOGGER.info("remove_rows_after_last_match - Exception found: %s" % (e))
        print(e)
        return df
```

### convert BIOES tagged labels to BIO tagged labels

```python
def BIOES2BIO(input_labels_list):
    output_labels_list = []
    for labels in input_labels_list:
        new_labels = []
        sent_len = len(labels)
        for idx in range(sent_len):
            if "-" not in labels[idx]:
                new_labels.append(labels[idx])
            else:
                label_type = labels[idx].split('-')[-1]
                if "E-" in labels[idx]:
                    new_labels.append("I-" + label_type)
                elif "S-" in labels[idx]:
                    new_labels.append("B-" + label_type)
                else:
                    new_labels.append(labels[idx])
        assert len(labels) == len(new_labels)
        output_labels_list.append(new_labels)
    return output_labels_list
```

```python
def conlltags2tree(sentence, chunk_types=('NP','PP','VP'),
                   root_label='S', strict=False):
    """
    Convert the CoNLL IOB format to a tree.
    """
    tree = Tree(root_label, [])
    for (word, postag, chunktag, prob, box, segment_id, ocr_conf) in sentence:
        if chunktag is None:
            if strict:
                raise ValueError("Bad conll tag sequence")
            else:
                # Treat as O
                tree.append((word, postag, prob, box, segment_id, ocr_conf))
        elif chunktag.startswith('B-'):
            tree.append(Tree(chunktag[2:], [(word, postag, prob, box, segment_id, ocr_conf)]))
        elif chunktag.startswith('I-'):
            if (len(tree)==0 or not isinstance(tree[-1], Tree) or
                tree[-1].label() != chunktag[2:]):
                if strict:
                    raise ValueError("Bad conll tag sequence")
                else:
                    # Treat as B-*
                    tree.append(Tree(chunktag[2:], [(word, postag, prob, box, segment_id, ocr_conf)]))
            else:
                tree[-1].append((word, postag, prob, box, segment_id, ocr_conf))
        elif chunktag.startswith('O-'):
            tree.append((word, postag, prob, box, segment_id, ocr_conf))
        else:
            raise ValueError("Bad conll tag %r" % chunktag)
    return tree
```

```python
# tag each token with pos
# Reference link - https://medium.com/analytics-vidhya/bio-tagged-text-to-original-text-99b05da6664
import numpy as np
pos_tags = [pos for token, pos in pos_tag(tokens)]
# convert the BIO / IOB tags to tree
conlltags = [(token, pos, tg, pb, box, segment_id, ocr_prob) for token, pos, tg, pb, box, segment_id, ocr_prob in zip(tokens, pos_tags, tags, prob, actual_box, text_segment_ids, ocr_score)]
ne_tree = conlltags2tree(conlltags)
# parse the tree to get our original text
original_text = []
for subtree in ne_tree:
    # checking for 'O' tags
    if type(subtree) == Tree:
        original_label = subtree.label()
        original_string = " ".join([token for token, pos, prob, box, segment_id, ocr_prob in subtree.leaves()])
        probabilty_score = np.mean([float(prob) for token, pos, prob, box, segment_id, ocr_prob in subtree.leaves()])
        actual_box = [box for token, pos, prob, box, segment_id, ocr_prob in subtree.leaves()]
        segment_ids = [segment_id for token, pos, prob, box, segment_id, ocr_prob in subtree.leaves()]
        ocr_probabilities = np.mean([float(ocr_prob) for token, pos, prob, box, segment_id, ocr_prob in subtree.leaves()])
        # print(original_label, original_string, actual_box, segment_ids[0])
        if (original_label, original_string, probabilty_score, actual_box, segment_ids[0], ocr_probabilities) not in original_text:
          original_text.append((original_label, original_string, probabilty_score, actual_box, segment_ids[0], ocr_probabilities))

# print(dict(original_text))
```

```python
import pdfplumber
from difflib import SequenceMatcher as SM
from nltk.util import ngrams
import codecs

def fuzzy_similarty_match(pdf_path, page_num, query):
    with pdfplumber.open(pdf_path) as pdf:
      page_num = pdf.pages[page_num]
      pdf_text = page_num.extract_text()

    needle = query
    hay = pdf_text
    needle_length  = len(needle.split())
    max_sim_val    = 0
    max_sim_string = u""

    for ngram in ngrams(hay.split(), needle_length + int(.2*needle_length)):
        hay_ngram = u" ".join(ngram)
        similarity = SM(None, hay_ngram, needle).ratio()
        if similarity > max_sim_val:
            max_sim_val = similarity
            max_sim_string = hay_ngram
    return (max_sim_val, max_sim_string)


def ocr_post_processing(result, field_list, pdf_path):
    ocr_threshold_value = 0.90
    ner_threshold_value = 0.90
    for key, value in result.items():
        value[0]['ocr_confidence'] = float(value[0]['ocr_confidence'])/100
        ocr_confidence = round(float(value[0]['ocr_confidence']), 4)
        ner_confidence = round(float(value[0]['ner_confidence']), 4)
        minimum_conf = min(ocr_confidence, ner_confidence)
        if ocr_confidence < ocr_threshold_value:
            if key in field_list:
                max_sim_val, max_sim_string = fuzzy_similarty_match(pdf_path, int(value[0]['page_number']), value[0]['value'])
                # print(max_sim_val, max_sim_string)
                if max_sim_val >= 0.9:
                    value[0]['value'] = max_sim_string
                    value[0]['overall_confidence'] = max_sim_val
                    value[0]['ManualQCRequired'] = False
                else:
                    value[0]['overall_confidence'] = ocr_confidence
                    value[0]['ManualQCRequired'] = True
            else:
                value[0]['overall_confidence'] = ocr_confidence
                value[0]['ManualQCRequired'] = True

        elif ner_confidence < ocr_confidence and ner_confidence < ner_threshold_value:
            value[0]['overall_confidence'] = ner_confidence
            value[0]['ManualQCRequired'] = True
        else:
            value[0]['overall_confidence'] = minimum_conf
            value[0]['ManualQCRequired'] = False
    return result
```
